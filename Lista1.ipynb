1) Plote um gráfico de T por N, onde T é o tempo médio que demora para conseguirmos pelo
menos N caras consecutivas em uma sequência de lançamentos consecutivos de moedas

```python
def consecutivas(N, p=[0.5, 0.5]):
    moedas = []
    consec = 0
    
    while consec < N:
        moeda = np.random.choice([0,1], p=p) # 0 == cara e 1 == coroa
        moedas.append(moeda)
        if moeda == 0:
            consec += 1
        else:
            consec = 0
           
    T = len(moedas)
    
    return T, moedas

def tempo_medio(N, t=10**3):
    ts = list(map(lambda x: consecutivas(x)[0], [N for i in range(0, t)]))
    return sum(ts) / len(ts)
```

```python
# Vou colocar n_max = 10, pois aparentemente é exponencial e está demorando muito!!!!!
n_max = 10
xs = [i for i in range(0, n_max+1)]
ys = list(map(lambda x: tempo_medio(x), xs))
```

```python
fig, ax = plt.subplots()

ax.plot(xs, ys)
ax.set(xlim=(0, n_max), ylim=(0, max(ys)))
plt.ylabel("T (Tempo Médio Até Conseguir >= N Caras)")
plt.xlabel("N (N Caras Consecutivas)")
plt.show()
```

3) Suponha que você tem N moedas. Você lança todas elas aleatoriamente e elas saem como
cara ou coroa. Então, você seleciona apenas as que saíram como coroa e as lança
novamente. Depois do lançamento, você seleciona as que saíram coroa e repete o processo
até que todas saiam como cara. Simule o número esperado de lançamentos em função de N.

```python
def todas_coroas(N, p=[0.5, 0.5]):
    n_caras = 0
    n_coroas = 0
    N_jogadas = N
    
    T = 0
    
    while n_caras != N:
        
        moedas = [np.random.choice([0,1], p=p) for i in range(0, N_jogadas)] # 0 == cara e 1 == coroa
        n_coroas = moedas.count(1)
        n_caras = len(moedas) - n_coroas
        T += len(moedas)
        
        moedas = [np.random.choice([0,1], p=p) for i in range(0, n_coroas)]
        n_coroas = moedas.count(1)
        n_caras = len(moedas) - n_coroas
        T += len(moedas)
        N_jogadas = n_coroas
        
    return T       
```

4) Mostre por simulação o curioso resultado de que, para um par de filhos A e B, a
probabilidade de que ambos sejam do sexo masculino dado que um deles é do sexo masculino
não é 50%.

```python
def simulacao(N=10**7):
    sexos = ['M', 'F']
    soma = 0
    
    for i in range(0, N):
        A, B = (random.choice(sexos), random.choice(sexos))
        while A != 'M' and B != 'M':
            A, B = (random.choice(sexos), random.choice(sexos))
        
        if A == 'M' and B == 'M':
            soma += 1
        
    return soma / N

simulacao()
```

5) Uma questão interessante em teoria de probabilidade é o problema do aniversário, que
versa sobre a probabilidade de 2 pessoas de um grupo fazerem aniversário em um mesmo dia
do ano. a) Plote um gráfico de P por N simulando a probabilidade P de 2 pessoas quaisquer
fizeram aniversário no mesmo dia em um grupo de N pessoas. b) Plote o mesmo gráfico para 3
pessoas.

```python
def msm_niver(N, p=2, bissexto=False):
    dias = 365 if bissexto is False else 366
    nivers = [np.random.randint(1, dias+1) for i in range(0, N)]
    
    unicos = set(nivers)
    d = dict([(u, nivers.count(u)) for u in unicos])
    
    return any(map(lambda d: d >= p, d.values()))

def probabilidade(N, p=2, bissexto=False, t=10**3):
    mc = [msm_niver(N, p, bissexto) for i in range(0, t)]
    
    return mc.count(True) / len(mc) 
```

6) Assuma que um bêbado está abraçado a um poste no ponto de coordenada (0,0). Considere
que o bêbado decide ir para casa, mas em seu estado inebriado ele caminha 1 metro em em
uma direção aleatória a cada passo. a) Plote o histograma do número de passos que o bêbado
leva para alcançar uma distância maior que 5 metros em relação ao poste. b) Plote um gráfico
D por N, mostrando a distância média D que o bêbado está do poste após N passos.

```python
def bebado(inicial=(0,0), p=[0.25, 0.25, 0.25, 0.25]):
    passo = random.choices([(-1, 0), (0, -1), (1, 0), (0, 1)], weights=p, k=1)[0]    
    return (inicial[0] + passo[0], inicial[1] + passo[1])   
```

```python
# (A)

def n_passos(inicial=(0,0), dm=5, p=[0.25, 0.25, 0.25, 0.25]):
    
    dist = lambda a, b: (abs(a[0] - b[0])**2 + abs(a[1] - b[1])**2)**(1/2)
    atual = inicial
    n = 0
    
    while dist(inicial, atual) <= dm:
        atual = bebado(atual, p)
        n +=1
        
    return n

N=10**5
xs = []
for i in range(0, N):
    xs.append(n_passos())
```

```python
# (A)

fig, ax = plt.subplots()

ax.hist(ys)
plt.xlabel("N (Número de Passos até Distância = 5m)")
plt.ylabel("H(N) (Frequência de N Passos até Distância = 5m)")
plt.show()
```

```python
# (B)

def distancia_n(N, inicial=(0,0), p=[0.25, 0.25, 0.25, 0.25]):
    
    dist = lambda a, b: (abs(a[0] - b[0])**2 + abs(a[1] - b[1])**2)**(1/2)
    atual = inicial
    
    while N > 0:
        atual = bebado(atual, p)
        N -= 1
        
    return dist(inicial, atual)

def media_n(N, inicial=(0,0), p=[0.25, 0.25, 0.25, 0.25], iters=10**5):
    soma = 0
    for i in range(0, iters):
        soma += distancia_n(N, inicial, p)
    
    return soma / iters

def plot_N_by_D(max_passos, inicial=(0,0), p=[0.25, 0.25, 0.25, 0.25]):
    xs = [i for i in range(0, max_passos+1)]
    ys = list(map(lambda x: media_n(x, p=p), xs))
    
    fig, ax = plt.subplots()

    ax.plot(xs, ys)
    plt.xlabel("N (Número de Passos)")
    plt.ylabel(f"D(N) (Distância Média do Início {str(inicial)} após N Passos)")
    plt.show()

plot_N_by_D(30)
```

7) A figura do final da lista mostra a América do Sul como um grafo de adjacências de países.
Suponha um processo markoviano em que, a cada passo, um turista viaja de um país para
outro do grafo. A cada passo, o turista sai do país atual e tem probabilidade igual de escolher
qualquer um dos países adjacentes como destino. a) Qual a probabilidade de, após um número
grande de passos, o turista estar no Brasil? b) Quantos passos em média o turista precisa dar
para visitar pelo menos uma vez cada país do grafo? c) Qual a probabilidade de um turista que
está agora no Brasil estar ainda no Brasil daqui 2, 3 ou 4 passos? d) Qual a probabilidade que
em uma caminhada de 10 passos começando no Brasil o turista não visite o Chile?

```python
def viagem(G, inicio="Br", N=10):
    adjs = lambda pais: list(filter(lambda x: x[0] == pais or x[1] == pais, G[1]))
    
    while N > 0:
        a,b = random.choice(adjs(inicio))
        if a == inicio:
            inicio = b
        else:
            inicio = a
            
        N -= 1
            
    return inicio
```

```python
G = (['E', 'Co', 'Pe', 'V', 'Br', 'G', 'S', 'FG', 'Bo', 'Pa', 'A', 'U', 'Ch'],
     [("E", "Co"), ("E", "Pe"), ("Co", "Pe"), ("Co", "V"), ("Co", "Br"),
      ("V", "Br"), ("V", "G"), ("G", "Br"), ("G", "S"), ("S", "FG"),
      ("S", "Br"), ("FG", "Br"), ("Br", "Pe"), ("Br", "Bo"), ("Br", "Pa"),
      ("Br", "A"), ("Br", "U"), ("Pe", "Bo"), ("Pe", "Ch"), ("Bo", "Ch"),
      ("Bo", "A"), ("Bo", "Pa"), ("Pa", "A"), ("Ch", "A"), ("A", "U")]
    )
```

```python
# (A)
def probabilidade_A(G, inicio="Br", N=10**3):
    soma = 0
    
    for i in range(0, N):
        if viagem(G, inicio, N) == "Br":
            soma += 1
            
    return soma / N
            
probabilidade_A(G)
```

```python
# (B)
def media_passos_B(G, inicio="Br", N=10**5):
    ps = G[0]
    passos = 0
    
    for i in range(0, N):
        visitados = [inicio]
        passos_rodada = 0
        atual = inicio
        
        while not all(map(lambda p: p in visitados, ps)):
            prox = viagem(G, atual, 1)
            passos_rodada += 1
            atual = prox
        
            if atual not in visitados:
                visitados.append(atual)
            
        passos += passos_rodada
            
    return passos / N

media_passos_B(G)
```

```python
# (C)
def probabilidade_C(G, inicio="Br", N=10**5):
    soma = 0
    
    for i in range(0, N):
        atual = viagem(G, inicio, 2)
        
        if atual == "Br":
            soma += 1
            continue
            
        atual = viagem(G, atual, 1)
        if atual == "Br":
            soma += 1
            continue
            
        atual = viagem(G, atual, 1)
        if atual == "Br":
            soma += 1
            
    return soma / N

probabilidade_C(G)
```

```python
# (D)
def probabilidade_D(G, inicio="Br", fim="Ch", N=10**5):
    soma = 0    
    
    for i in range(0, N):
        visitados = [inicio]
        atual = inicio
        
        for j in range(0, 10):
            atual = viagem(G, atual, 1)
            visitados.append(atual)
            
        if fim not in visitados:
            soma += 1
            
    return soma / N

probabilidade_D(G)
```

(“prisão”), simule novamente as probabilidades. c) Considerando que as melhores
propriedades imobiliárias são aquelas onde os jogadores param com frequência, qual você
diria que é a melhor cor de propriedades no tabuleiro?

```python
def nova_posicao(inicial=0, p1=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], p2=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6]):
    d1 = random.choices([1,2,3,4,5,6], weights=p1)[0]
    d2 = random.choices([1,2,3,4,5,6], weights=p2)[0]
    
    return (inicial + d1 + d2) % 40
```

```python
propriedades = [(1,'Leblon', 'Rosa'), (3, 'Av. Presidente Vargas', 'Rosa'), (4, 'Av. Nossa S. de Copacabana', 'Rosa'),
(6, 'Av. Brigadeiro Faria Lima', 'Azul'), (8, 'Av. Rebouças', 'Azul'), (9, 'Av. 9 de Julho', 'Azul'),
(11, 'Av. Europa', 'Roxo'), (13, 'Rua Augusta', 'Roxo'), (14, 'Rua Pacaembú', 'Roxo'),
(17, 'Interlagos', 'Laranja'), (19, 'Morumbi', 'Laranja'),
(21, 'Flamengo', 'Vermelho'), (23, 'Botafogo', 'Vermelho'),
(26, 'Av. Brasil', 'Amarelo'), (28, 'Av. Paulista', 'Amarelo'), (29, 'Jardim Europa', 'Amarelo'),
(31, 'Copacabana', 'Verde'), (33, 'Av. Vieira Souto', 'Verde'), (34, 'Av. Atlântica', 'Verde'), (36, 'Ipanema', 'Verde'),
(38, 'Jardim Paulista', 'Azul'), (39, 'Brooklin', 'Azul')
]
```

```python
# (A)

def prob_prop(inicial=0, props=propriedades, p1=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6],
              p2=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], N=10**7):
    
    ps = [inicial]
    
    for i in range(0, N):
        inicial = nova_posicao(inicial)
        ps.append(inicial)
    
    res = []
    for pp, np, cp in props:
        res.append((pp, np, cp, ps.count(pp) / N))
        
    return res

# Sem nenhuma outra informação (cadeia, "volte para tal lugar", etc) todas são equiprováveis
prob_prop()  
```

```python
# (B)

def prob_prop_com_cadeia(inicial=0, props=propriedades, p1=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6],
                         p2=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6], N=10**7):
    
    va_para_prisao = 30
    prisao = 10
    
    ps = [inicial]
    
    for i in range(0, N):
        inicial = nova_posicao(inicial)
        
        if inicial == va_para_prisao:
            inicial = prisao
        
        ps.append(inicial)
    
    res = []
    for pp, np, cp in props:
        res.append((pp, np, cp, ps.count(pp) / N))
        
    return res
    
com_cadeia = prob_prop_com_cadeia()
com_cadeia
```

9) a) Os times de futebol Vermelho FC e Azul FC jogam uma partida. Suponha que os times
marcam gols de forma independente seguindo processos de Poisson com taxas de chegada λv
= 2/partida e λa = 3/partida. Qual a taxa de partidas que podemos esperar que o time Vermelho
FC ganha? b) Resolva o mesmo exercício para os times de basquete RedCats e BlueDogs,
assumindo taxas de chegada de pontos de λr = 100/partida e λb = 150/partida, respondendo
quantas vezes podemos esperar que o time RedCats ganha. c) Compare os resultados e
explique.

```python
# (A)

def taxa_vermelho(lv=2, la=3, N=10**7):
    vence = 0
    
    for i in range(0, N):
        gv = np.random.poisson(lv)
        ga = np.random.poisson(la)
        
        if gv > ga:
            vence += 1
            
    return vence / N

taxa_vermelho()
```

```python
# (B)

def taxa_redcats(lr=100, lb=150, N=10**7):
    vence = 0
    
    for i in range(0, N):
        gr = np.random.poisson(lr)
        gb = np.random.poisson(lb)
        
        if gr > gb:
            vence += 1
            
    return vence / N

taxa_redcats()
```

```python
# (C)

# Em um primeiro momento, poderíamos pensar que as taxas de (A) e (B) seriam
# iguais, uma vez que a razão entre as chegadas (razão dos lambdas) são igual-
# mente 2/3 (2/3 e 100/150 = 2/3), porém, devemos pensar que, na verdade, no
# primeiro caso, a comparação é que o Azul tem somente 1 de taxa de chegada
# (gol) a frente do Vermelho, fazendo que, dada essa diferença de somente 1,
# ele ganhará 75% das vezes. Agora, já no caso do Blue vs Red, o Blue faz 50
# cestas a frente do Red, ou seja, tem muito mais probabilidade de fazer mais
# cestas que o Red --- fazendo, dessa forma, que o Blue saia vencedor em vir-
# tualmente todo jogo.
```
